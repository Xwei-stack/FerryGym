```python
# load dataset from pkl file
import pandas as pd
import numpy as np

# load dataset from pkl file
df = pd.read_pickle('./ferrygym/2022-04-10-13->14.pkl')
```

```python
# replace each mmsi with a unique integer
mmsi_list = df['mmsi'].unique()
mmsi_dict = {mmsi_list[i]: i for i in range(len(mmsi_list))}
df['mmsi'] = df['mmsi'].map(mmsi_dict)
```

```python
# only keep columns datetime, mmsi, x, y, direction
df = df[['datetime', 'mmsi', 'x', 'y', 'direction']]

# save as dataset.pkl
df.to_pickle('./ferrygym/dataset.pkl')
```

```python
# shipA = [[1184.01171121, 2252.01597619]
#  [1209.68523548, 2221.34236943]
#  [1217.35363717, 2227.7607505 ]
#  [1191.6801129, 2258.43435726]]
# shipB = [[1152.90897559, 2268.98354651]
#  [1183.64031654, 2243.37915853]
#  [1190.04141353, 2251.06199377]
#  [1159.31007259, 2276.66638175]]

#  plot shipA and shipB as polygons and check if they intersect
import matplotlib.pyplot as plt
import matplotlib.path as mpltPath

shipA = np.array([[1184.01171121, 2252.01597619],
 [1209.68523548, 2221.34236943],
 [1217.35363717, 2227.7607505 ],
 [1191.6801129,  2258.43435726]])

shipB = np.array([[1152.90897559, 2268.98354651],
 [1183.64031654, 2243.37915853],
 [1190.04141353, 2251.06199377],
 [1159.31007259, 2276.66638175]])

pathA = mpltPath.Path(shipA)
pathB = mpltPath.Path(shipB)

fig, ax = plt.subplots()
ax.plot(shipA[:,0], shipA[:,1], 'r')
ax.plot(shipB[:,0], shipB[:,1], 'b')
plt.show()

print(pathA.intersects_path(pathB))
```
